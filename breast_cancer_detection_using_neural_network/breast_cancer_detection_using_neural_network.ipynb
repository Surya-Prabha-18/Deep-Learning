```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
# scaled_array = sc.fit_transform(data_frame)
# scaled_df = pd.DataFrame(scaled_array, columns=data_frame.columns)
```

```python
# for i in data_frame.columns:
#     data_frame[[i]] = sc.fit_transform(data_frame[[i]])
```

```python
x = data_frame.drop('label', axis = 1)
y = data_frame['label']
```

```python
# for i in x.columns:
#     x[[i]] = sc.fit_transform(x[[i]])
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x_train_std = sc.fit_transform(x_train)
x_test_std = sc.fit_transform(x_test)
```

```python
import tensorflow as tf
from tensorflow import keras
```

**flatten** -> change n dimensional array to 1d array
**dense** - > helps in connection of all nodes to all nodes or layers

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)),
    keras.layers.Dense(20, activation = 'relu'),
    keras.layers.Dense(2, activation = 'sigmoid')
])
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(x_train_std, y_train, validation_split=0.1, epochs = 20)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['training_data', 'validation_data'], loc = 'lower right')
plt.ylabel('accutacy')
plt.xlabel('epoch')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['training_data', 'val_acccuracy'], loc = 'lower right')
```

```python
y_pred = model.predict(x_test_std)
```

```python
y_pred
```

```python
y_pred1 = np.argmax(y_pred, axis = 1)
```
