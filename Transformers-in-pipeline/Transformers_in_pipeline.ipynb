```python
question_answer = pipeline("question-answering")
```

```python
question_answer(
    question = "Which ocean is the largest on Earth?",
    context = "The Pacific Ocean is the largest and deepest ocean on Earth. It covers more than 63 million square miles and reaches depths of over 36,000 feet in the Mariana Trench. It borders many countries including the United States, Japan, and Australia. The ocean plays a vital role in regulating the Earth's climate and is home to diverse marine life.",
)
```
